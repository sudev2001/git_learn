```python
murder = df.loc[0,['Cases Pending Investigation from Previous Year - Col. (3)','Cases Reported during the year - Col. (4)','Cases Reopened for Investigation - Col. (5)','Total Cases for Investigation (Col.3+Col.4+Col.5) - Col. (6)','Cases Not Investigated Under 157_1_b CRPC - Col. (7)','Cases Transferred to Other State or Agency - Col. (8)']]
hit_and_run = df.loc[4,['Cases Pending Investigation from Previous Year - Col. (3)','Cases Reported during the year - Col. (4)','Cases Reopened for Investigation - Col. (5)','Total Cases for Investigation (Col.3+Col.4+Col.5) - Col. (6)','Cases Not Investigated Under 157_1_b CRPC - Col. (7)','Cases Transferred to Other State or Agency - Col. (8)']]
```

```python
murder
hit_and_run
explode=[0]*len(murder)
explode[3]=.125
```

```python
labels=['Previous Year','This Year','Cases Reopened','Total Cases','Not Investigated','Cases Transferred']
plt.pie(murder,labels=labels,wedgeprops={'edgecolor':'black'},shadow=True,autopct='%1.1f%%',explode=explode,startangle=90)
plt.tight_layout()
plt.show()
murder
```

```python
pl_x=['Previous Year','This Year','Cases Reopened','Total Cases','Not Investigated','Cases Transferred']
pl_y = murder
py_y = hit_and_run
plt.figure(figsize=(10,5))
plt.plot(pl_x,pl_y,color='k',ls='--',marker='o',label='Murder Report')
plt.plot(pl_x,py_y,color='b',ls=':',marker='*',label='Hit and Run')
plt.legend()
plt.title('Case Study')
plt.tight_layout()
plt.show()
```

```python
pl_x=['Previous Year','This Year','Cases Reopened','Total Cases','Not Investigated','Cases Transferred']
pl_y = murder
py_y = hit_and_run
plt.figure(figsize=(10,5))
# plt.yticks(range(0, max(py_y)+100, 10))
width=.25
x_index = np.arange(len(pl_x))
plt.bar(x_index,pl_y,label='Murder',width=width)
plt.bar(x_index+width,py_y,label='Hit and Run',width=width)
plt.xticks(x_index,pl_x)
plt.legend()
plt.title("CASE STUDY")
plt.tight_layout()
plt.show()
```
